```python
x=housing_dropped.isna().sum()
housing_null = pd.DataFrame({"Column": x.index,\
                             "number of null values": x.values})
housing_null
```

The DOM column, which indicated the number of days the ad was placed on the site, has a lot of missing values, and if we want to delete the columns whose DOM value is not available, almost half of our data will be deleted! On the other hand, this feature has important consequences for houses, and removing the DOM column is not a sensible thing to do.

We can use the column mean or mode to fill in the blank values. Let's take a look at its box diagram, maybe it will help us in making a decision.

```python
housing_dropped.DOM.plot(kind = 'box')
```

as you see there is lots of outbound values. Therefore, using mode is better option

```python
housing_no_missing = housing_dropped.copy()
housing_no_missing.loc[housing_no_missing["DOM"].isna() ,"DOM"]=housing_no_missing["DOM"].mode()[0]
housing_no_missing

housing_no_missing.dropna(axis=0 , inplace=True)

housing_no_missing
```

```python
housing_no_missing.info()
```

After performing this operation, we no longer have lost data and can go to the next step. In the next step, we will use the "housing_no_missing" data frame. Pay attention that when saving the file, we must choose an encoder that is able to recognize all the characters of the data frame.

```python
housing_no_missing.to_csv('housing_no_missing.csv', encoding='gbk', index=False)
```
